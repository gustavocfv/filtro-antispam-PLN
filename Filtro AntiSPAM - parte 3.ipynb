# 2.Carregar classificador Naive  Bays

````python
import pickle
````

````python
def carregar_modelo(nome_arquivo):
    nome = str(nome_arquivo) + '.pickle'
    try:
        arquivo = open(nome, "rb")
        modelo = pickle.load(arquivo)
        arquivo.close()
        return modelo
    except Exception as e:
        return e
````

````python
classificador = carregar_modelo('naivebayes')
````

````python
classificador
````

# 4.Classificar todas as mensagens

````python
df_classificados = pd.DataFrame(columns=['SMS', 'LABEL', 'prob_ok', 'prob_blocked'])
df_classificados
````

````python
from time import time
````

````python
ti = time()

for i in range(len(dados_sem_classificação)):
    if i % 1000 == 0:
        print(i, end=" ")
    
    frase = processamento_NLP_utils.estruturar_dados([dados_sem_classificação[i]])
    resultado = processamento_NLP_utils.classificar_texto(classificador, frase[0][0])
 
    df_classificados.loc[i] = [dados_sem_classificação[i][0],
                               resultado['classe'],
                               resultado['ok'],
                               resultado['blocked']
                              ]
tf = time() - ti
print("\nTempo total {} em segundos".format(tf))
````
